### Cleaning Data 

### Change `start_time` and `end_time` datatype to datetime

```python
# change data type of (start_time, end_time ) to datetime
start_time=pd.to_datetime(Bikes.start_time)
end_time=pd.to_datetime(Bikes.end_time)
```

```python
print("The data type of (start_time) is {} and the data type of (end_time) is {} .".format(start_time.dtype, end_time.dtype))
```

### Check for duplicates

```python
# Check for duplicates
print( "Count of duplicated values: ", sum((Bikes.duplicated())))
```

### Drop unneeded columns

```python
# code for dropping unneeded columns
bikes=Bikes.drop(['start_station_latitude','start_station_longitude','end_station_longitude','end_station_latitude'],axis=1)
```

```python
# Test for the new dataframe
print("The exist columns after dropping unneeded ones are: \n\n {} \n \n The new dataframe has {} rows and {} columns\
".format(bikes.columns, len(bikes), bikes.shape[1]))
```

### Drop all rows with NAN values for all Columns 

```python
# Code to count all rows with NAN Values
df1 = bikes[bikes.isna().any(axis=1)]
print("Count of all rows with NAN values are {}: ",df1.shape[0])
```

```python
# Drop all rows with NANs
bikes.dropna(inplace=True)
```

```python
#Test if there any NANs
bikes.isna().sum()
```

### Change `member_birth_year` datatype to integer

```python
# code to change datatype of member_birth_year to integer
member_birth_year=bikes.member_birth_year.astype(int)
```

```python
# test the new datatype of member_birth_year
member_birth_year.dtype
```

### Add new column `age` to our dataframe

```python
year=bikes.member_birth_year.values
age=2019-year
bikes['age']=age.astype('int')
bikes.info()
```

## Univariate Exploration

## First: Plotting categorical variables

### Plotting gender:
### First: with absolute values:

```python
# count for each gender
gender_count=bikes.member_gender.value_counts()

# resize the figure size
plt.figure(figsize=[8,5]);

# Choose the first tuple of RGB colors
base_color = sb.color_palette()[0]


sb.countplot(data=bikes, x='member_gender', color=base_color, order=gender_count.index);

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=0) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = gender_count[label.get_text()]
    pct_string = '{:0}'.format(count)

    # print the annotation just below the top of the bar
    plt.text(loc, count, pct_string, ha = 'center', color = 'blue',fontsize=14)
    
# plot view
plt.title("Count of users according to their Gender",fontsize=16);
plt.xlabel("Gender",fontsize=14);
plt.ylabel("number of users",fontsize=14);
plt.xticks(fontsize=12);
plt.yticks(fontsize=12);
```

### Plotting gender:
### Second: with relative values:

```python
# count of all observations
observations_count=len(bikes)

# resize the figure size
plt.figure(figsize=[8,5]);


plt.yticks(fontsize=12);

sb.countplot(data=bikes, x='member_gender', color=base_color, order=gender_count.index);

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=0) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = gender_count[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/observations_count)

    # print the annotation just below the top of the bar
    plt.text(loc, count, pct_string, ha = 'center', color = 'blue',fontsize=14)

# plot view
plt.title("proportion of users according to their Gender",fontsize=16);
plt.xlabel("Gender",fontsize=14);
plt.ylabel("number of users",fontsize=14);
plt.xticks(fontsize=12);
```

> ##### We can figure out that the proportion of males is around `75%` of all observations

### Plotting Age

```python
# resize the figure size
plt.figure(figsize=[8,8]);
bins = np.arange(0, bikes['age'].max()+1,1)
plt.hist(data = bikes, x = 'age', bins = bins,rwidth = 0.7, color='#0504aa');
plt.xlim(0, 100,10);
plt.title("Distribution of age",fontsize=16);
plt.xlabel("age",fontsize=14);
plt.ylabel("number of users",fontsize=14);
plt.xticks(fontsize=12);
plt.yticks(fontsize=12);
```

### Plotting trip duration

```python
binsize = 500
bins = np.arange(0, bikes['duration_sec'].max()+binsize, binsize)

plt.figure(figsize=[8, 8])
plt.hist(data = bikes, x = 'duration_sec', bins = bins,rwidth = 0.7, color='#0504aa')
plt.title('Distribution of Trip Durations', fontsize=16)
plt.xlabel('Duration in seconds', fontsize=14)
plt.ylabel('Number of Trips', fontsize=14)
plt.axis([-200, 10000, 0, 90000])
plt.show()
```

>### This data that, in their natural units, can look highly skewed: lots of points with low values, with a very long tail of data points with large values. So let's put it on a log scale instead

```python
log_binsize = 0.05
bins_log = 10 ** np.arange(2.4, np.log10(bikes['duration_sec'].max()) + log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = bikes, x = 'duration_sec', bins = bins_log, rwidth = 0.7, color='#0504aa');
plt.title('Distribution of Trip Durations', fontsize=16);
plt.xlabel('Duration in seconds',fontsize=14);
plt.ylabel('Number of Trips',fontsize=14);
plt.xscale('log');
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k']);
plt.axis([0, 10000, 0, 15000]);
plt.show();
```

#### Most of trips takes duration in betwwen 100 and 1000 seconds

## Multivariate Exploration

### First: Plotting `age`, `gender`, and `trip duration`:

```python
gender_markers = [['Male', 'o'],['Female', '^'],['Other', 's']]

for gender, marker in gender_markers:
    df_gender = bikes[bikes['member_gender'] == gender]
    plt.scatter( df_gender['age'], df_gender['duration_sec'], marker = marker, alpha= 1);
plt.title("Plotting Member Gender according to Age and Trip duration ",fontsize=16)
plt.xlim([10, 100])
plt.ylim([0, 10000])
plt.xlabel("Age", fontsize=14)
plt.ylabel("Trip Duration in seconds", fontsize=14)
plt.legend(['Male','Female','Other'])
plt.show()
```

> #### I think that the plot isn't clear enough, so I need to separate it into three plots to show each category of gender separately

```python
gender = sb.FacetGrid(data = bikes, col = 'member_gender',xlim = [10, 80], ylim = [0, 10000],\
                      margin_titles=True,col_wrap = 3, height=5);
gender.map(plt.scatter, 'age', 'duration_sec');
gender.set_xlabels('Age',fontsize=14);
gender.set_ylabels('Trip Duration in seconds', fontsize=14);
plt.show();
```

> #### We can notice that:
> - Number of `Males` is higher than number of `Females`.
> - Trip duration for both `gender` is decreasing by age.
> - For older users (age between `60` and `70`) `males` take higher trip duration than `females`.

### Second: Plotting `user type`, `gender`, and `trip duration`:

```python
UserType_markers = [['Subscriber', 'o'],['Customer', '^']]

for User_type, marker in UserType_markers:
    df_User_type = bikes[bikes['user_type'] == User_type]
    plt.scatter( df_User_type['age'], df_User_type['duration_sec'], marker = marker, alpha= 1);
plt.title("Plotting User Type according to Age and Trip duration ",fontsize=16)
plt.xlim([10, 100])
plt.ylim([0, 10000])
plt.xlabel("Age", fontsize=14)
plt.ylabel("Trip Duration in seconds", fontsize=14)
plt.legend(['Subscriber','Customer'])
plt.show()
```

> #### I think that the plot isn't clear enough, so I need to separate it into two plots to show each category of user type separately
